```python
text = ''
for sentences in range(len(dataset)):
    text += dataset.iloc[sentences][0]
```

```python
text = text.lower()  
text = re.sub('[^a-z]', ' ', text)  
text = re.sub(r'\s+', ' ', text)
```

```python
all_text = nltk.sent_tokenize(text) 
text_list = [nltk.word_tokenize(sent) for sent in all_text]
print(text_list)
for i in range(len(text_list)):  
    text_list[i] = [word for word in text_list[i] if word not in stopwords.words('english')]
```

```python
model = gensim.models.Word2Vec(size=300, window=5, min_count=2, workers=4)
```

```python
model.build_vocab(text_list)
```

```python
model.intersect_word2vec_format('testing_data/GoogleNews-vectors-negative300.bin', lockf=1.0, binary=True)
```

```python
model.train(text_list, total_examples=model.corpus_count, epochs=model.iter)
```
