```python
import pandas as pd
from spellchecker import SpellChecker
from gensim.models import Word2Vec
```

```python
def init_model(model):
    return Word2Vec.load(model) 
```

```python
def check_en(word, model):
    spell = SpellChecker()
    word = word.lower()
    i = 0
    sim = 0
    highest_sim = []
    try:       
        while i < 5:
            sim_word = model.wv.most_similar([str(word)], topn=5000)[sim][0]
            sim += 1
            inner_check_text = spell.unknown([sim_word])
            if len(inner_check_text) == 0:
                if sim_word.isdigit() != True:
                    highest_sim.append(sim_word)
                    i += 1
    except Exception as e:
        highest_sim.append(("Word is not in thesaurus"))
    return highest_sim
```

```python
testing_dict_sg = dict()
testing_dict_cbow = dict()
word = ''
ts_model_sg = init_model("../newtrained_model/tsdict_w2v_skipgram.model") 
ts_model_cbow = init_model("../newtrained_model/tsdict_w2v_cbow.model") 
testing_wordlist = pd.read_csv('../testing_use/testing_list.csv', low_memory=False)
for textspeak in range(len(testing_wordlist)):
    word = testing_wordlist.iloc[textspeak][0]        
    testing_dict_sg[word] = check_en(word, ts_model_sg)
    testing_dict_cbow[word] = check_en(word, ts_model_cbow)
```
